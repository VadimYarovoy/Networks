```python
import numpy as np
import matplotlib.pyplot as plt

#    (1  0)
bits = [1, 0, 1, 0, 1, 0, 1, 0]

#   
sample_rate = 1000  #  
bit_duration = 1  #   
num_samples_per_bit = int(sample_rate * bit_duration)

#         
half_period = 0.67
t = np.linspace(0, bit_duration, num_samples_per_bit)
sinusoid = np.sin(2 * np.pi * t / half_period)
reverse_sinusoid = -sinusoid

#     
signal = []

for bit in bits:
    if bit == 1:
        signal.extend(sinusoid)
    else:
        signal.extend(reverse_sinusoid)

#  
t_signal = np.linspace(0, len(signal) / sample_rate, len(signal))
plt.plot(t_signal, signal, linewidth=2)
plt.xlabel("")
plt.ylabel("")
plt.title("    1  0")
plt.ylim(-1.1, 1.1)
plt.grid(True)
plt.show()
```
