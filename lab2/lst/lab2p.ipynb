```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем массив битов (1 и 0)
bits = [1, 0, 1, 0, 1, 0, 1, 0]

# Создаем параметры сигнала
sample_rate = 1000  # Частота дискретизации
bit_duration = 1  # Продолжительность каждого бита
num_samples_per_bit = int(sample_rate * bit_duration)

# Создаем сигнал синусоиды и обратной синусоиды для полтора периода
half_period = 0.67
t = np.linspace(0, bit_duration, num_samples_per_bit)
sinusoid = np.sin(2 * np.pi * t / half_period)
reverse_sinusoid = -sinusoid

# Создаем массив значений для графика
signal = []

for bit in bits:
    if bit == 1:
        signal.extend(sinusoid)
    else:
        signal.extend(reverse_sinusoid)

# Создаем график
t_signal = np.linspace(0, len(signal) / sample_rate, len(signal))
plt.plot(t_signal, signal, linewidth=2)
plt.xlabel("Время")
plt.ylabel("Значение")
plt.title("График сигнала с битами 1 и 0")
plt.ylim(-1.1, 1.1)
plt.grid(True)
plt.show()
```
