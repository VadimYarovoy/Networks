```julia
import Pkg
Pkg.add("DataFrames")
Pkg.add("CSV")
```

```julia
using DataFrames
using CSV
```

```julia
function do_stuff(filename)
    df = CSV.read(filename, DataFrame)
    array = Float64[]

    for row in eachrow(df)
        push!(array, round(row[6] / row[7],  digits=6))
    end

    df.Err = array

    df
end
```

```julia
file_names = ["datasets/test-05.csv", "datasets/test-01.csv", "datasets/test-005.csv"]
results = []
for file in file_names
    res = do_stuff(file)
    push!(results, res)
    res |> println
end
```

```julia
for df in results 
    for row in eachrow(df)
        println(
            "[\$", row[1],"\$], ",
            "[\$", row[2],"\$], ",
            "[\$", row[3],"\$], ",
            "[\$", row[4],"\$], ",
            "[\$", row[5],"\$], ",
            "[\$", row[6],"\$], ",
            "[\$", row[7],"\$], ",
            "[\$", row[8],"\$], ",
        )
    end
    println()
end
```
